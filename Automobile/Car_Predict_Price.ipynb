```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test= train_test_split(x,
                                                   y,
                                                   test_size=0.20,
                                                   random_state=0)
```

```python
print('x_train shape:', x_train.shape)
print('x_test shape:', x_test.shape)
print('y_train shape:', y_train.shape)
print('y_test shape:', y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression

Car_Price = LinearRegression()
Car_Price.fit(x_train, y_train)
```

```python
m=Car_Price.coef_
c=Car_Price.intercept_
m,c
```

```python
c
```

```python
y_pred_train = Car_Price.predict(x_train)
y_pred_train
```

```python
y_pred_test = Car_Price.predict(x_test)
y_pred_test
```

```python
from sklearn.metrics import mean_absolute_error,r2_score
```

```python
mean_absolute_error(y_test,y_predict_test)
```

```python
mean_absolute_error(y_train,y_predict_train)
```

```python
from sklearn.metrics import mean_squared_error,r2_score
```

```python
mean_squared_error(y_test,y_predict_test)
```

```python
mean_squared_error(y_train,y_predict_train)
```

```python
accuracy=r2_score(y_test,y_predict_test)
```

```python
accuracy
```

```python
print("accuracy is",accuracy*100)
```

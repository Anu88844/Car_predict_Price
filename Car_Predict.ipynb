```python

from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test= train_test_split(x,
                                                   y,
                                                   test_size=0.20,
                                                   random_state=0)
```

```python
print('x_train shape:', x_train.shape)
print('x_test shape:', x_test.shape)
print('y_train shape:', y_train.shape)
print('y_test shape:', y_test.shape)
```

```python
#import LinearRegression
from sklearn.linear_model import LinearRegression
```

```python
#using LinearRegression in New Dataset
Car_Price=LinearRegression()
```

```python
#fit x_train & y_train in new_dataset
Car_Price.fit(x_train,y_train)
```

```python
m = Car_Price. coef_
c = Car_Price.intercept_
m,c
```

```python
c
```

```python
#predict x_train
y_predict_train = Car_Price.predict(x_train)
y_predict_train
```

```python
#predict x_test
y_predict_test = Car_Price.predict(x_test)
y_predict_test
```

```python
#import mean_absolute_error
from sklearn.metrics import mean_absolute_error,r2_score
```

```python
#checking mean_absolute_error between y_test & y_predict_test 
mean_absolute_error(y_test,y_predict_test)
```

```python
#checking mean_absolute_error between y_train & y_predict_train
mean_absolute_error(y_train,y_predict_train)
```

```python
#import mean_squared_error 
from sklearn.metrics import mean_squared_error,r2_score
```

```python
#checking mean_squared_error between y_test & y_predict_test
mean_squared_error(y_test,y_predict_test)
```

```python
#checking mean_squared_error between y_train & y_predict_train
mean_squared_error(y_train,y_predict_train)
```

```python
#checking accuracy between y_test & y_predict_test
accuracy=r2_score(y_test,y_predict_test)
```

```python
#print aacuracy
accuracy
```

```python
#print accuracy in percentage
print("accuracy is",accuracy*100)
```

```python
#coefficient of determination between y_train & y_predict_train
r2score_train= r2_score(y_train, y_predict_train)
r2score_train
```

```python
#using regplot for prediction price
plt.style.use("default")

plt.figure(figsize=(4, 4))
sns.regplot(x=y_test, y=y_predict_test, ci=None, color="red")
```

```python
#using scatter plot between actual point & predict point
plt.figure(figsize=(4,4))
plt.style.use("ggplot")

plt.scatter(x=range(10),
            y=y_test.head(10),
            color="r",
           label="Actual Points")
plt.scatter(x=range(10),
            y=y_predict_test[:10],
            color="g",
           label="Predicted Points")
plt.legend()
```
